## 1.2 What is Hypothesis Testing?

Hypothesis testing is a statistical method used to determine whether a specific claim (hypothesis) about a population is true or not. In this process, sample data is analyzed to provide insights into population parameters. Hypothesis testing typically involves comparing two main hypotheses:

**Null Hypothesis (H0):** Represents the current state or assumption.

**Alternative Hypothesis (H1):** Proposes a claim contrary to the null hypothesis.

The test is conducted using a significance level (usually 5% or 0.05), which indicates the risk of accepting a false result.

## 1.3 What is Assumption Check?

Before performing hypothesis testing, it is essential to check whether certain assumptions required by the test are satisfied. These checks ensure the reliability and validity of the test results. Examples of common assumptions include:

**Normal Distribution:** Whether the sample data follows a normal distribution.

**Equality of Variances:** Whether variances across groups are equal.

**Independence:** Whether observations are independent of each other.

Skipping assumption checks can lead to misleading results.

## 1.4 What is the Null Hypothesis (H0)?

The null hypothesis assumes that the current condition is true. It serves as the baseline to disprove. For example:

-Claims there is no difference between two groups.

-States that a population parameter equals a specific value.

**Example:** In a drug study, the null hypothesis might be, "The effectiveness of this drug is equal to the placebo."

## 1.5 What is the Alternative Hypothesis (H1)?

The alternative hypothesis represents a claim that contradicts the null hypothesis. It is accepted when the null hypothesis is rejected. This hypothesis suggests that there is a change, difference, or relationship.

**Example:** "The effectiveness of this drug is not equal to the placebo."

## 1.6 How to Select an Appropriate Hypothesis Test?

Choosing the right hypothesis test depends on the type of data and the research question. The following steps can help guide the selection:

**Identify the data type:** Is the dependent variable continuous or categorical?

**Check the number of groups:** Are you comparing one, two, or more groups?

**Perform assumption checks:** Are assumptions like normality and equality of variances met?

**Define the research goal:** Are you testing for differences, relationships, or proportions?

**Examples of Test Types:**

-t-Test: Compares the means of two groups.

-ANOVA: Compares the means of more than two groups.

-Chi-Square Test: Analyzes the relationship between categorical variables.

-Regression Analysis: Models the relationship between variables.

# 3 Assumption Check

## 3.1 Normality Check
**Purpose:** To determine whether the data follows a normal distribution, which is a common assumption in many statistical tests (e.g., t-tests, ANOVA).

**Test Used:** Shapiro-Wilk Test.

-Null Hypothesis (H‚ÇÄ): The data is normally distributed.

 -Alternative Hypothesis (H‚ÇÅ): The data is not normally distributed.

**Interpretation of p-value:**

-If p-value ‚â• 0.05, we fail to reject H‚ÇÄ and conclude the data is normal.

-If p-value < 0.05, we reject H‚ÇÄ and conclude the data is not normal.

```python
def check_normality(data):
    test_stat_normality, p_value_normality=stats.shapiro(data)
    print("p value:%.4f" % p_value_normality)
    if p_value_normality <0.05:
        print("Reject null hypothesis >> The data is not normally distributed")
    else:
        print("Fail to reject null hypothesis >> The data is normally distributed")

```

## 3.2 Variance Homogeneity Check
**Purpose:** To test whether two or more groups have equal variances, which is an assumption for certain statistical tests (e.g., t-tests, ANOVA).

**Test Used:** Levene's Test.

-Null Hypothesis (H‚ÇÄ): The variances of the groups are equal.

-Alternative Hypothesis (H‚ÇÅ): The variances of the groups are different.

**Interpretation of p-value:**

-If p-value ‚â• 0.05, we fail to reject H‚ÇÄ and conclude that the variances are equal.

-If p-value < 0.05, we reject H‚ÇÄ and conclude that the variances are different.


```python
def check_variance_homogeneity(group1, group2):
    test_stat_var, p_value_var= stats.levene(group1,group2)
    print("p value:%.4f" % p_value_var)
    if p_value_var <0.05:
        print("Reject null hypothesis >> The variances of the samples are different.")
    else:
        print("Fail to reject null hypothesis >> The variances of the samples are same.")
```

## 5.1 McNemar Testing

A company wants to evaluate the effectiveness of a new training program in improving employee knowledge. They give a pre-test and a post-test to the same group of 100 employees. The results are binary: "Pass" or "Fail". The company uses the McNemar Test, which is appropriate for paired categorical data, to determine if the training program significantly improved the pass rate.

### 5.1.1 Test Selection

Tests the change in proportions for paired categorical data (e.g., success rates before and after a new training program).

### 5.1.2 Hypotheses

H‚ÇÄ (Null Hypothesis): The training program has no effect on pass rates. (The proportions of "Pass" and "Fail" do not change between pre-test and post-test.)

H‚ÇÅ (Alternative Hypothesis): The training program has an effect on pass rates. (The proportions of "Pass" and "Fail" change between pre-test and post-test.)

```python
table = np.array([[30, 20],
                  [40, 10]])

result = mcnemar(table, exact=False, correction=True)
print("Test Statistic:", result.statistic)
print("P-value:", result.pvalue)

if result.pvalue < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```
